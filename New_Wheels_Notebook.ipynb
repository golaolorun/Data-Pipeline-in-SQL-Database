# New Wheels Project Visualization


### Context

A lot of people in the world share a common desire: to own a vehicle. A car or an automobile is seen as an object that gives the freedom of mobility. Many are now preferring pre-owned vehicles because they come at an affordable cost, but at the same time, they are also concerned about whether the after-sales service provided by the resale vendors is as good as the care you may get from the actual manufacturers. New-Wheels, a vehicle resale company, has launched an app with an end-to-end service from listing the vehicle on the platform to shipping it to the customer's location. This app also captures the overall after-sales feedback given by the customer. 

### Problem Statement:

New-Wheels sales have been dipping steadily in the past year, and due to the critical customer feedback and ratings online, there has been a drop in new customers every quarter, which is concerning to the business. The CEO of the company now wants a quarterly report with all the key metrics sent to him so he can assess the health of the business and make the necessary decisions.


### Objective
As a data scientist, you see that there is an array of questions that are being asked at the leadership level that needs to be answered using data. But this data is not organized and is just being dumped as flat files to be used only when there are concerning problems! You need to first create a pipeline to organize and maintain this data using a SQL database so that it becomes easy to answer questions in the future. Once these questions have been answered by writing SQL queries, you can use those results to create the Quarterly Business Report. Use the results from the queries* to generate visuals in Python using matplotlib or seaborn libraries. 


### Let us start by importing the required libraries

```python
# import libraries for data manipulation
import numpy as np
import pandas as pd

# import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

### Understanding the structure of the data

```python
# read the the tables generated from querying the data in SQL to answer each business question
df1 = pd.read_csv('Question1.csv')
df2 = pd.read_csv('Question2.csv')
df3 = pd.read_csv('Question3.csv')
df4 = pd.read_csv('Question4.csv')
df5 = pd.read_csv('Question5.csv')
df6 = pd.read_csv('Question6.csv')
df7 = pd.read_csv('Question7.csv')
df8 = pd.read_csv('Question8.csv')
df9 = pd.read_csv('Question9.csv')
df10 = pd.read_csv('Question10.csv')
```

### Question 1: What is the distribution of customers across states

```python
#View top five rows
df1.head()
```

```python
# What is the distribution of customers across states
sns.catplot(x='STATE', y='no_of_customers', data=df1, kind="bar", height=5, aspect=2)
plt.xticks(rotation = 90)
plt.show()
```

### Question 2: What is the average rating in each quarter?

```python
#View top five rows
df2.head()
```

```python
#What is the average rating in each quarter?
sns.catplot(x='QUARTER_NUMBER', y='AVERAGE_RATINGS', data=df2, kind="bar", height=6, aspect=2)
plt.show()
```

### Question 3: Are customers getting more dissatisfied over time?

```python
#View top five rows
df3.head()
```

```python
#Are customers getting more dissatisfied over time?
sns.catplot(x='CUSTOMER_FEEDBACK', y='FEEDBACK_PERCENTAGE', data=df3, kind="bar", height=6, aspect=2)
plt.show()
```

### Question 4: Which are the top 5 vehicle makers preferred by the customer.

```python
#View top five rows
df4.head()
```

```python
#Which are the top 5 vehicle makers preferred by the customer.
sns.catplot(x='VEHICLE_MAKER', y='No_of_Cust', data=df4, kind="bar", height=5, aspect=2)
plt.xticks(rotation = 90)
plt.show()
```

### Question 5: What is the most preferred vehicle make in each state?

```python
#View top five rows
df5.head()
```

```python
#What is the most preferred vehicle make in each state?
plt.figure(figsize = (20,10))     #Visualize number of most preferred cars with a Bar Graph
sns.countplot(data = df5, x = 'VEHICLE_MAKER')
plt.xticks(rotation=90);
plt.ylabel("Number_of_Preferred_State")
```

### Question 6: What is the trend of number of orders by quarters?

```python
#View top five rows
df6.head()
```

```python
#What is the trend of number of orders by quarters?
plt.figure(figsize = (10,4))
x=df6.QUARTER_NUMBER
y=df6.No_of_Orders
plt.plot(x,y)
plt.minorticks_on()
plt.xlabel("QUARTER_NUMBER")
plt.ylabel("No_of_Orders")
plt.grid()
plt.tick_params(labelsize=10, width=2)
plt.xticks([1,2,3,4])
plt.show()
```

### Question 7: What is the quarter over quarter % change in revenue? 

```python
#View top five rows
df7.head()
```

```python
#What is the quarter over quarter % change in revenue? 
plt.figure(figsize=(15,7))
sns.lineplot(data=df7, x='QUARTER_NUMBER' ,y='QUARTER OVER QUARTER REVENUE (%) CHANGE')
plt.tick_params(labelsize=10, width=2)
plt.xticks([2,3,4])
plt.yticks([-50,-40,-30,-20,-10,0,10,20,30])
plt.show()
```
